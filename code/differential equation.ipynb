# 放射性蜕变 & 复利

$u' = ku  $  即u在每一个瞬时的变化率与该瞬时的u的数值成比例。

这个微分方程的解是$u = u_0 e^{kt}$

$\frac{u_2}{u_1} = \frac{u_0 e^{kt_2}}{u_0 e^{kt_1}} = e^{k(t_2 - t_1)}$


```python
t=np.linspace(0,5,500)
u0 = 5
k = -0.5
u=u0*np.exp(k*t)
```

```python
fig = plt.figure(figsize=(9, 4),facecolor='white')
ax = fig.add_subplot(1, 2,1)
plt.plot(t,u,'r-')
plt.xlabel('$t$', fontsize = 20)
plt.ylabel('$u$', fontsize = 20)

ax = fig.add_subplot(1, 2,2)
plt.plot(t,u,'r-')
plt.xlabel('$t$', fontsize = 20)
plt.ylabel('$log(u)$', fontsize = 20)
plt.yscale('log')

plt.tight_layout()
plt.show()
```
